# Winnter take all assignemnt of cortical regions to cerebellar regions

## Working memory dataset:
* using hierarchical parcellation
* focus on D regions

```python
df = swta.assign_winner(cortex_roi = "glasser", #Icosahedron42
                        dataset = "WMFS",  
                        ses_id = "ses-02", 
                        type = "CondAll", 
                        roi_sup = "D", 
                        cerebellum_roi = "NettekovenSym68c32",
                        atlas_cerebellum = "SUIT3")
```

### do X2 test of independence
H0: cortical assignment using WTA and cerebellar roi are independent (randomly assigned?)


```python
p = swta.randomization_test(df, 
                           rows= "cereb_roi", 
                           shuffle = "assigned_num",
                           numIterations=100,
                           sides=2, 
                           nbins = 10)
```
